```python
from __future__ import division
import sympy
import numpy
%matplotlib inline
from scipy.integrate import odeint
from matplotlib import pyplot as plt
sympy.init_printing(use_latex='mathjax')

T=(0,36,65,100,160)
Ca=[0.1823,0.1453,0.1216,0.1025,0.0795]
t=numpy.linspace(0,170,100)
Co=[0.1823,0,55]
Kc=148.323
k=0.0067
m=1
n=1
o=1

def dCadt(C,t):
    return (-k*(C[0]**n)+(k/Kc)*(C[1]**m)*(C[2])**o,
           k*(C[0]**n)-(k/Kc)*(C[1]**m)*(C[2]**o),
            k*(C[0]**n)-(k/Kc)*(C[1]**m)*(C[2]**o))
 

Caint=odeint(dCadt,Co,T)
plt.figure(1)
plt.plot(T,Ca,'ro',T,Caint[:,0],'b')
plt.ylabel('Ca')
plt.xlabel('Time')
plt.show()
```

```python
#question2

T=[0,41,48,55,75,96,127,146,162,180,194,212,267,318,368,379,410]
Cc=[0,1.18,1.38,1.63,2.24,2.75,3.31,3.76,3.81,4.11,4.31,4.45,4.86,5.15,5.32,5.35,5.42]
t =numpy.linspace(0,410,200)

Cao=5.5
Cbo=5.5
Cco=0
Conc=[Cao,Cbo,Cco]
Kc=4.9763
k=0.00067
def dCcdt(C,t):
    return (-k*C[0]*C[1]+(k/Kc)*(C[2])**2,
            -k*C[0]*C[1]+(k/Kc)*(C[2])**2,
            2*(k*C[0]*C[1]-(k/Kc)*(C[2])**2))
        

Ccint=odeint(dCcdt,Conc,t)
plt.figure(1)
plt.plot(T,Cc,'ro',T,Ccint[:,2],'b')
plt.ylabel('Cc')
plt.xlabel('Time')
plt.show()
```

```python
#question5
Rr=[0.0087,0.0154,0.0254,0.0322,0.0323,0.0221,0.0093,-0.0013,-0.0077,-0.0108,-0.0121]
T=[0,8,16,24,32,40,48,56,64,72,80]
t=numpy.linspace(0,80,100)
C0=[2,0.1]
k1=0.05
k2=0.005
Ccat=3

def dCdt(C,t):
    return (-k1*C[0]*C[1],
           k1*C[0]*C[1]-k2*C[1]*Ccat)

Ccint=odeint(dCdt,C0,t)
Cr=Ccint[:,1]
Ca=Ccint[:,0]
rr=k1*Ca*Cr-k2*Cr*Ccat

plt.figure(1)
plt.plot(T,Rr,'ro',t,rr,'b')
plt.ylabel('Rr')
plt.xlabel('Time')
plt.show()
```
