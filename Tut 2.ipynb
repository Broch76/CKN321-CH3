```python
#question2
from __future__ import division
import sympy
import numpy
%matplotlib inline
from scipy.integrate import odeint
from matplotlib import pyplot as plt
sympy.init_printing(use_latex='mathjax')

Pt = numpy.array([1,1.5,1.65,1.76,1.84,1.9,1.95,1.99,2.025,2.08,2.12,2.15,2.175])
T = [0,0.5,1,1.5,2,2.5,3,3.5,4,5,6,7,8]
Po=1.3
Co=101.325/(8.314*287)
Cao=0.7694*Co
Cio=0.2306*Co

X= ((Pt*Co/Po)-Co)/Cao
k=2.5
n=1.5
def dXdt(X,T):
    return (k*(Cao*(1-X[0]))**n)/Cao

Xint=odeint(dXdt,0,T)

plt.figure(1)
plt.plot(T,X,'ro',T,Xint[:,0],'b')
plt.ylabel('Conversion')
plt.xlabel('Time')
plt.show()
```
